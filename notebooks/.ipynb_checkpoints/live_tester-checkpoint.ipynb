```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# TensorNetwork class
class TensorNetwork:
    def __init__(self, graph, tensors):
        self.graph = graph
        self.tensors = tensors
        self.index_dims = {}
        self.index_to_tensors = {}

        for name, (tensor, indices) in tensors.items():
            for idx, dim in zip(indices, tensor.shape):
                if idx in self.index_dims:
                    assert self.index_dims[idx] == dim, f"index {idx} has inconsistent dimension"
                else:
                    self.index_dims[idx] = dim
                self.index_to_tensors.setdefault(idx, []).append((name, tensor, indices))

# evaluate configuration
def evaluate_config(network, config):
    result = 1.0
    for name, (tensor, inds) in network.tensors.items():
        key = tuple(config[i] for i in inds)
        result *= tensor[key]
    return result

# Glauber update
def update_edge(network, config, idx, beta=1.0):
    tensors = network.index_to_tensors[idx]
    dim = network.index_dims[idx]
    probs = np.ones(dim)
    for name, arr, inds in tensors:
        slc = [config[i] if i != idx else slice(None) for i in inds]
        probs *= arr[tuple(slc)] ** beta
    probs = probs / probs.sum() if probs.sum() > 0 else np.ones(dim) / dim
    config[idx] = np.random.choice(dim, p=probs)

# contraction estimate with live plot
def estimate_contraction(network, betas, iters=10000, burns=1000, n_rounds=5, verbose=False, live_plot=False):
    logZ_factor = 0.0
    logZ_trajectory = []
    all_weights_per_beta = []

    if live_plot:
        plt.ion()
        fig, ax = plt.subplots(figsize=(10, 5))
        line, = ax.plot([], [], color='blue')
        ax.set_xlim(0, len(betas))
        ax.set_ylim(-10, 10)  # adjust as needed
        ax.set_xlabel('Beta Step')
        ax.set_ylabel('Cumulative log(Z)')
        ax.set_title('Live log(Z) Convergence')

    for i in range(1, len(betas)):
        beta_prev, beta_curr = betas[i - 1], betas[i]
        delta_beta = beta_curr - beta_prev
        all_weights = []

        for round_num in range(n_rounds):
            config = {idx: np.random.randint(dim) for idx, dim in network.index_dims.items()}
            round_weights = []

            for t in range(iters):
                idx = np.random.choice(list(network.index_dims.keys()))
                update_edge(network, config, idx, beta=beta_curr)
                if t >= burns:
                    w = evaluate_config(network, config)
                    if w > 0:
                        round_weights.append(w ** (delta_beta))
            all_weights.append(round_weights)

        final_log_prods = [np.sum(np.log(weights)) for weights in all_weights if len(weights) > 0]
        max_log = np.max(final_log_prods)
        ratio_estimate_log = max_log + np.log(np.mean(np.exp(final_log_prods - max_log)))
        logZ_factor += ratio_estimate_log

        if verbose:
            print(f'[beta={beta_curr:.3f}]  ratio: {ratio_estimate_log:.4e}  cumulative logZ: {logZ_factor:.4e}')

        logZ_trajectory.append(logZ_factor)
        all_weights_per_beta.append([np.std(w) / np.mean(w) if np.mean(w) > 0 else 0 for w in all_weights])

        if live_plot:
            line.set_data(range(len(logZ_trajectory)), logZ_trajectory)
            ax.relim()
            ax.autoscale_view()
            plt.pause(0.01)

    if live_plot:
        plt.ioff()
        plt.show()

    logZ_estimate = logZ_factor + np.sum(np.log(list(network.index_dims.values())))
    Z_estimate = np.exp(logZ_estimate)
    return Z_estimate, logZ_trajectory, all_weights_per_beta

# run multiple chains with live updates
def run_multiple_chains(network, betas, n_chains=5, iters=10000, burns=1000, n_rounds=5, Z_true=None):
    estimates = []
    logZ_trajectories = []
    weight_variances = []

    for seed in range(n_chains):
        np.random.seed(seed)
        print(f'Running chain {seed+1}/{n_chains}...')
        Z_est, logZ_traj, all_weights = estimate_contraction(
            network, betas, iters, burns, n_rounds, verbose=True, live_plot=True
        )
        estimates.append(Z_est)
        logZ_trajectories.append(logZ_traj)
        weight_variances.append([np.mean(var) for var in all_weights])

    # final plots: logZ trajectories
    plt.figure(figsize=(12, 4))
    for i, log_traj in enumerate(logZ_trajectories):
        plt.plot(betas[1:], log_traj, label=f'chain {i+1}', alpha=0.8)
    plt.xlabel('Beta')
    plt.ylabel('Cumulative log(Z)')
    plt.title('Final log(Z) Trajectories Across Chains')
    plt.legend()
    plt.grid(True)
    plt.show()

    # variance of weights
    plt.figure(figsize=(12, 4))
    for i, var in enumerate(weight_variances):
        plt.plot(betas[1:], var, label=f'chain {i+1}', alpha=0.8)
    plt.xlabel('Beta')
    plt.ylabel('Relative Std of Weights')
    plt.title('Variance of Weights Across Chains')
    plt.legend()
    plt.grid(True)
    plt.show()

    # relative error plot
    meanz = np.mean(estimates)
    stdz = np.std(estimates)
    final_Z_var = 0.0
    if meanz != 0:
        final_Z_var = stdz / meanz
    else:
        final_Z_var = float('nan')
    print(f'Relative variance of final Z estimates: {final_Z_var:.4f}')
    if Z_true is not None:
        errors = [abs(Z - Z_true) / abs(Z_true) for Z in estimates]
        plt.figure(figsize=(7, 5))
        plt.bar(range(1, n_chains + 1), errors, color='steelblue', alpha=0.7)
        plt.xlabel('Chain Number')
        plt.ylabel('Relative Error')
        plt.title('Relative Error of Final Z Estimates')
        plt.axhline(0, color='red', linestyle='--')
        plt.grid(True)
        plt.show()

    return np.mean(estimates), np.std(estimates)
```

```python
# test on 2x2 ring (Tr(ABCD))
def test_trace_ABCD(dim=3, betas=np.linspace(0, 1, 100), n_chains=5, iters=10000, burns=900, n_rounds=5):
    # create 2x2 ring tensor network
    G = nx.Graph()
    G.add_edges_from([
        ('A', 'B'), ('B', 'C'),
        ('C', 'D'), ('D', 'A'),
    ])

    # random positive tensors
    A, B, C, D = [np.random.rand(dim, dim) + 0.1 for _ in range(4)]
    tensors = {
        'A': (A, ['i', 'j']),
        'B': (B, ['j', 'k']),
        'C': (C, ['k', 'l']),
        'D': (D, ['l', 'i'])
    }

    # true value
    true_trace = np.trace(A @ B @ C @ D)

    # build network
    tn = TensorNetwork(G, tensors)

    # run multiple MCMC chains with live plots
    mean_Z, std_Z = run_multiple_chains(
        tn, betas, n_chains=n_chains, iters=iters, burns=burns,
        n_rounds=n_rounds, Z_true=true_trace
    )

    # summary
    print("\n================== Final Summary ==================")
    print(f"True Tr(ABCD): {true_trace:.6f}")
    print(f"Estimated Tr(ABCD): {mean_Z:.6f} ± {std_Z:.6f}")
    print(f"Relative error: {abs(mean_Z - true_trace) / abs(true_trace):.4%}")
    print("====================================================")

# run the test
test_trace_ABCD()
```
